```python
x = var('x')
f(x) = x + 10000

fig = plot(f, (x, 0, 6.5), ymin=0, ymax=6.5, gridlines=True, axes_labels=['', ''], figsize=(6, 6), thickness=2, ticks=[
         1, 1], gridlinesstyle=dict(color="#a20b6b", linestyle="-", linewidth=1)) 

points = []

for i in range(1, 7):
    for j in range(1, 7):
        points.append((i, j))


fig += point(points, size=93)
fig.fontsize(21)
fig.set_legend_options(font_size=21)

fig
```
